```python
import matplotlib.pyplot as plt
import random
import math
```

```python
X1 = [0,0,1,1]
X2 = [0,1,0,1]
#OR
Y1 = [0,1,1,1]
#XOR
Y2 = [0,1,1,0]
```

```python
plt.figure()

for i in range(len(Y1)):
  if Y1[i] == 0:
    plt.plot(X1[i], X2[i], "ro")
  else:
    plt.plot(X1[i],X2[i], "go")

plt.show()
```

```python
W = [random.random(), random.random()]
b = random.random()

print(W,b)

lr = 0.01
theta = 0.5
```

```python
# for OR
epoch = 0
count = 0

while count < 4:
  total_error = 0
  print(f"Epoch: {epoch}")
  for i in range(len(X1)):
    weighted_sum = W[0]*X1[i] + W[1]*X2[i] + b
    pred = 0
    if weighted_sum > theta:
      pred = 1

    error = Y1[i] - pred
    if error !=0:
      count = 0
      total_error += math.sqrt(error**2)

      W[0] += lr*error*X1[i]
      W[1] += lr*error*X2[i]
      b += lr*error
    else:
      count +=1

    print(f"[{X1[i]}, {X2[i]}] = {pred}")

  print(f"Loss: {total_error}")
  print("------------------------------")
  epoch += 1
```

```python
#for XOR
epoch = 0
count = 0

while count < 4:
  total_error = 0
  print(f"Epoch: {epoch}")
  for i in range(len(X1)):
    weighted_sum = W[0]*X1[i] + W[1]*X2[i] + b
    pred = 0
    if weighted_sum > theta:
      pred = 1

    error = Y2[i] - pred
    if error !=0:
      count = 0
      total_error += math.sqrt(error**2)

      W[0] += lr*error*X1[i]
      W[1] += lr*error*X2[i]
      b += lr*error
    else:
      count +=1

    print(f"[{X1[i]}, {X2[i]}] = {pred}")

  print(f"Loss: {total_error}")
  print("------------------------------")
  epoch += 1
```
